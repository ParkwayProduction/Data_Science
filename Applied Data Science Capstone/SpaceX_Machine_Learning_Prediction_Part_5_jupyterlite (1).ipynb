This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix');
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed'])
    plt.show()
```

## TASK  12


Find the method performs best:


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import jaccard_score, f1_score, log_loss, accuracy_score

# Example Data Preparation
data = pd.DataFrame({
    'Feature1': np.random.rand(100),
    'Feature2': np.random.rand(100),
    'Class': np.random.randint(0, 2, 100)
})

X = data[['Feature1', 'Feature2']]
Y = data['Class']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train different models
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)

tree = DecisionTreeClassifier()
tree.fit(X_train, Y_train)

forest = RandomForestClassifier()
forest.fit(X_train, Y_train)

# List of predictors and their names
predictors = [logreg, tree, forest]
predictor_names = ['Logistic Regression', 'Decision Tree', 'Random Forest']

# Check the unique values in Y_test
unique_labels = np.unique(Y_test)
print(f"Unique labels in Y_test: {unique_labels}")

# Assuming '1' represents 'yes' and '0' represents 'no'
pos_label = 1 if 1 in unique_labels else unique_labels[0]

# Initialize the best results and corresponding predictors
best_predictor = ""
best_accuracy = 0
best_f1_score = 0
best_jaccard_index = 0
best_log_loss = float('inf')

# Dictionary to store all results
results = {}

for predictor, name in zip(predictors, predictor_names):
    y_pred = predictor.predict(X_test)
    y_pred_proba = predictor.predict_proba(X_test) if hasattr(predictor, "predict_proba") else None

    accuracy = accuracy_score(Y_test, y_pred)
    f1 = f1_score(Y_test, y_pred, pos_label=pos_label)
    jaccard = jaccard_score(Y_test, y_pred, pos_label=pos_label)
    logloss = log_loss(Y_test, y_pred_proba) if y_pred_proba is not None else None

    # Store the results
    results[name] = {
        "Accuracy": accuracy,
        "F1-Score": f1,
        "Jaccard Index": jaccard,
        "LogLoss": logloss
    }

    # Update best predictor based on accuracy
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_predictor = name

    # Update best predictor based on F1-Score
    if f1 > best_f1_score:
        best_f1_score = f1
        best_predictor_f1 = name

    # Update best predictor based on Jaccard Index
    if jaccard > best_jaccard_index:
        best_jaccard_index = jaccard
        best_predictor_jaccard = name

    # Update best predictor based on LogLoss
    if logloss is not None and logloss < best_log_loss:
        best_log_loss = logloss
        best_predictor_logloss = name

# Print all results
print("Model Evaluation Results:")
for name, metrics in results.items():
    print(f"\n{name}:")
    for metric, value in metrics.items():
        print(f"{metric}: {value}")

# Print the best predictors based on different metrics
print(f"\nBest Predictor based on Accuracy: {best_predictor} with a score of {best_accuracy}")
print(f"Best Predictor based on F1-Score: {best_predictor_f1} with a score of {best_f1_score}")
print(f"Best Predictor based on Jaccard Index: {best_predictor_jaccard} with a score of {best_jaccard_index}")
print(f"Best Predictor based on LogLoss: {best_predictor_logloss} with a score of {best_log_loss}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification

# Generate a more realistic dataset
X, Y = make_classification(n_samples=1000, n_features=20, n_informative=2, n_redundant=10, random_state=42)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train different models
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, Y_train)

svm = SVC(probability=True)
svm.fit(X_train, Y_train)

tree = DecisionTreeClassifier()
tree.fit(X_train, Y_train)

knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)

# List of predictors and their names
predictors = [logreg, svm, tree, knn]
predictor_names = ['Logistic Regression', 'SVM', 'Decision Tree', 'KNN']

# Dictionary to store accuracy results
accuracy_results = {}

for predictor, name in zip(predictors, predictor_names):
    y_pred = predictor.predict(X_test)
    accuracy = accuracy_score(Y_test, y_pred)
    accuracy_results[name] = accuracy

# Create a bar chart for accuracy
plt.figure(figsize=(10, 6))
plt.bar(accuracy_results.keys(), accuracy_results.values(), color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.ylim(0, 1)
for i, (name, accuracy) in enumerate(accuracy_results.items()):
    plt.text(i, accuracy + 0.01, f'{accuracy:.2f}', ha='center')
plt.show()
```
